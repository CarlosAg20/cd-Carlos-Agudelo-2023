```python
import matplotlib.pyplot as plt
```

```python
x = [76, 92, 107, 42, 65, 150, 65, 105, 40, 55]
y = [200, 320, 300, 160, 190, 315, 230, 380, 155, 175]

px = sum(x) / len(x)

py = sum(y) / len(y)

print(px, py)

sum_xy = sum(a * b for a, b in zip(x, y))

sum_xx = sum(a * a for a in x)

m = (sum_xy - len(x) * px * py) / (sum_xx - len(x) * px * px)
b = (py - m * px)

print(f"pendiente {m}\ncorte con eje, {b}" )
print(f"Función: y = 1.9x + 87")
```
